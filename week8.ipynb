### 实验要求

1. 使用pandas读取数据集github_bot_processed_data.csv。探索pandas的显示选项，以便查看更多数据行和列。


```python
import pandas as pd

data = pd.read_csv('github_bot_processed_data.csv')
print(data.head())

data.info()
print(data.describe())
```

2. 使用head()方法查看数据的前几行，并使用info()方法查看每列的数据类型，观察两者的区别。
使用describe()方法生成数据的描述性统计信息。


```python
data['created_at'] = pd.to_datetime(data['created_at'])

print(data.head())
```

3. 对数据进行格式化处理，并展示不同格式（例如，日期、货币、百分比等）的效果。

```python
import numpy as np

data['log_public_repos'] = np.log(data['public_repos'] + 1)
data['log_public_gists'] = np.log(data['public_gists'] + 1)
data['log_followers'] = np.log(data['followers'] + 1)
data['log_following'] = np.log(data['following'] + 1)

print(data[['log_public_repos', 'log_public_gists', 'log_followers', 'log_following']].head())
```

4. 对public_repos、public_gists、followers、following等列进行对数变换，并查看其影响。

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
data['label'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("类别分布")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()

bool_columns = ['site_admin', 'company']
data[bool_columns].sum().plot(kind='bar', stacked=True, color=['lightblue', 'salmon'])
plt.title("布尔特征堆积柱状图")
plt.xlabel("Boolean Feature")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(data['log_public_repos'], bins=30, color='skyblue', edgecolor='black')
plt.title("log_public_repos的分布")
plt.xlabel("log_public_repos")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(data['public_repos'], data['followers'], alpha=0.5, color='lightcoral')
plt.title("public_repos vs followers")
plt.xlabel("public_repos")
plt.ylabel("followers")
plt.show()

import seaborn as sns

sns.pairplot(data[['public_repos', 'public_gists', 'followers', 'following']])
plt.show()
```

5. 使用Matplotlib绘制图表
    - 绘制条形图：展示label列的类别分布。
    - 绘制堆积柱状图：展示多个布尔特征（如site_admin、company等）的分布。
    - 绘制直方图：展示log_public_repos的对数变换后的数据分布。
    - 绘制散点图：展示public_repos与followers之间的关系。
    - 绘制散点矩阵：展示多个数值型特征之间的成对关系。


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('github_bot_processed_data.csv')

plt.figure(figsize=(8, 6))
df['label'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Label Distribution')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

bool_columns = ['site_admin', 'company', 'blog', 'location', 'hireable']
df[bool_columns] = df[bool_columns].applymap(lambda x: 1 if x else 0)

df[bool_columns].sum().plot(kind='bar', stacked=True, color=['lightcoral', 'lightskyblue'])
plt.title('Distribution of Boolean Features')
plt.xlabel('Feature')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

df['log_public_repos'] = np.log(df['public_repos'] + 1)  
plt.figure(figsize=(8, 6))
plt.hist(df['log_public_repos'], bins=20, color='salmon', edgecolor='black')
plt.title('Log-Transformed Public Repos Distribution')
plt.xlabel('Log(Public Repos)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df['public_repos'], df['followers'], alpha=0.5, color='purple')
plt.title('Public Repos vs Followers')
plt.xlabel('Public Repos')
plt.ylabel('Followers')
plt.show()

num_columns = ['public_repos', 'followers', 'following', 'text_bot_count']
sns.pairplot(df[num_columns], kind='scatter')
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

6. 使用Seaborn绘制图表
    - 绘制箱线图：展示不同label类别下log_followers的分布。
    - 绘制成对图：展示不同特征之间的成对关系，并根据label分类。
    - 绘制热图：展示log_public_repos、log_public_gists、log_followers和log_following等特征之间的相关性。
    - 绘制小提琴图：展示label与log_followers之间的分布差异。


```python
import seaborn as sns

df['log_followers'] = np.log(df['followers'] + 1)

plt.figure(figsize=(8, 6))
sns.boxplot(x='label', y='log_followers', data=df, palette='Set2')
plt.title('Distribution of Log(Followers) by Label')
plt.xlabel('Label')
plt.ylabel('Log(Followers)')
plt.show()

sns.pairplot(df[['public_repos', 'followers', 'following', 'text_bot_count', 'label']], hue='label', palette='Set1')
plt.suptitle('Pairplot of Numerical Features by Label', y=1.02)
plt.show()

df['log_public_repos'] = np.log(df['public_repos'] + 1) 
df['log_public_gists'] = np.log(df['public_gists'] + 1) 
df['log_following'] = np.log(df['following'] + 1)  

correlation_matrix = df[['log_public_repos', 'log_public_gists', 'log_followers', 'log_following']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Log Transformed Features')
plt.show()

plt.figure(figsize=(8, 6))
sns.violinplot(x='label', y='log_followers', data=df, palette='Set2')
plt.title('Log(Followers) Distribution by Label')
plt.xlabel('Label')
plt.ylabel('Log(Followers)')
plt.show()
```

7. 使用pandas_profiling.ProfileReport()生成交互式数据分析报告，分析数据的统计分布、缺失值、异常值等。

```python
import pandas as pd
from pandas_profiling import ProfileReport

df = pd.read_csv('github_bot_processed_data.csv')

profile = ProfileReport(df, title='GitHub Bot Processed Data Report', explorative=True)

profile.to_file('github_bot_processed_data_report.html')

profile.to_notebook_iframe()
```
